```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
#
```

```python
df=pd.read_csv('metaads-performance.csv')
```

```python
df
```

```python
df['Date']=pd.to_datetime(df['Date'])
```

```python

df.isnull().sum()
```

```python
df.fillna(0, inplace=True)
```

```python
df.duplicated().sum()
```

```python
df['ROI (%)'] = (df['Revenue'] - df['Cost']) / df['Cost'] * 100
```

```python
df['profit']=df['Revenue']-df['Cost']
```

```python
df
```

```python
df['Conversion Rate (%)'] = (df['Conversions'] / df['Clicks']) * 100
```

```python
df
```

```python
df.describe()
```

```python
# Aggregate by month to see trends
df['month']=df['Date'].dt.month
monthly_data = df.groupby('month').agg({
    'Impressions': 'sum',
    'Reach': 'sum',
    'Cost': 'sum',
    'Clicks': 'sum',
    'Conversions': 'sum',
    'Revenue': 'sum',

}).reset_index()


```

```python
monthly_data
```

```python


# Calculate 'Conversion Rate (%)' if it doesn't exist
if 'Conversion Rate' not in df.columns:
  df['Conversion Rate '] = (df['Conversions'] / df['Clicks'])

# Calculate 'Profit' if it doesn't exist
if 'Profit' not in df.columns:
  df['Profit'] = df['Revenue'] - df['Cost']

# Calculate 'Return on Conversions'
df['Return on Conversions'] = df['Revenue'] / df['Conversions']

# Calculate 'ROI' if it doesn't exist
if 'ROI' not in df.columns:
  df['ROI'] = (df['Revenue'] - df['Cost']) / df['Cost']

df
```

```python


df = df.drop('Date', axis=1)
df
```

```python

aggregated_data = df.sum().to_frame().transpose()
aggregated_data
```

```python


df_subset = aggregated_data[[ 'Impressions', 'Clicks', 'Cost',
                 'Conversions', 'Conversion Rate (%)', 'Revenue', 'Profit',
                 'Return on Conversions', 'ROI']]
df_subset
```

```python
import matplotlib.pyplot as plt

# Plot Total Revenue by Date
plt.figure(figsize=(12, 6))
plt.plot(df['Date'], df['Revenue'], marker='o', linestyle='-', color='b')
plt.title('Total Revenue Over Time')
plt.xlabel('Date')
plt.ylabel('Revenue')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
correlation_matrix = df[['Impressions', 'Clicks', 'Cost', 'Conversions', 'Revenue', 'ROI (%)']].corr()
print(correlation_matrix)

# Plot Correlation Matrix
import seaborn as sns
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix')
plt.show()



```
