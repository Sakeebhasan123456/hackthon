```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df['profit']=df['Revenue']-df['Cost']
```

```python
df['month']=df['Date'].dt.month
df=df.drop('Date',axis=1)
```

```python
df
```

```python
aggregated_data = df.groupby(['Campaign type']).agg({
    'Impressions': 'sum',
    'Clicks': 'sum',
    'Cost': 'sum',
    'Conversions': 'sum',
    'Revenue': 'sum',


}).reset_index()
```

```python
aggregated_data
```

```python
aggregated_data.to_csv('microsoftbudget_data.csv', index=False)
```

```python
total_profit = df['profit'].sum()
total_cost = df['Cost'].sum()


roi_percentage = (total_profit / total_cost) * 100

print(f'Total ROI Percentage: {roi_percentage:.2f}%')
```

```python

aggregated_data['Label'] = aggregated_data['Campaign type'] + ' - Month ' + aggregated_data['month'].astype(str)

# Plotting
plt.figure(figsize=(10, 10))
plt.pie(aggregated_data['profit'],
        labels=aggregated_data['Label'],
        autopct='%1.1f%%',
        startangle=90,
        colors=sns.color_palette('viridis', len(aggregated_data)))

plt.title('Aggregate Profit Distribution by Campaign Type and Month')
plt.show()
```

```python

aggregated_data['Return on Conversions'] = aggregated_data['Revenue'] / aggregated_data['Conversions']


plt.figure(figsize=(8, 8))
plt.pie(aggregated_data['Return on Conversions'], labels=aggregated_data['Campaign type'], autopct='%1.1f%%', startangle=90, colors=sns.color_palette('viridis'))
plt.title('Return on Conversions by Campaign Type')
plt.show()
```

```python

aggregated_data['Conversion Rate (%)'] = (aggregated_data['Conversions'] / aggregated_data['Cost']) * 100
```
